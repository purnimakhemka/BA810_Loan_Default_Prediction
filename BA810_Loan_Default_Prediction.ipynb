### ***4. Computing Credit History***

*Note: Due to the large amount of Data in our primary dataset from here we will only be using `neny_loan_clean` as our primary dataset.*

#### ***4.1. Converting to Datetime***

<p align='justify'>We here are transforming columns with date into Date Time format to make them calculation ready for computing the Credit History at the Time of Loan Origination</p>

```python
# Work directly on neny_loan_clean
# (assumes 'issue_d' like '2019-03-01' and 'earliest_cr_line' like 'Dec-1990')
neny_loan_clean = neny_loan_clean.copy()

# Parse dates
neny_loan_clean['issue_d_dt'] = pd.to_datetime(
    neny_loan_clean['issue_d'].astype(str).str.strip(),
    errors='coerce'              # issue_d is already YYYY-MM-DD
)

neny_loan_clean['earliest_cr_line_dt'] = pd.to_datetime(
    neny_loan_clean['earliest_cr_line'].astype(str).str.strip(),
    format='%b-%Y',              # Mon-YYYY, e.g. "Dec-1990"
    errors='coerce'
)
```

#### ***4.2. Computing Credit History at the Time of loan Origination***

<p align='justify'>Here we are computing the Credit History at the time of loan origination and dropping the original two columns as they are not needed once we have the credt history column</p>

```python
# Computeing credit history at origination in years
neny_loan_clean['credit_history_years'] = (
    (neny_loan_clean['issue_d_dt'] - neny_loan_clean['earliest_cr_line_dt'])
    .dt.days / 365.25
)

# Handleing negatives and missing values
neny_loan_clean['credit_history_years'] = neny_loan_clean['credit_history_years'].clip(lower=0)

median_history = neny_loan_clean['credit_history_years'].median()
neny_loan_clean['credit_history_years'] = neny_loan_clean['credit_history_years'].fillna(median_history)

# Droping the original string date columns if no longer needed
neny_loan_clean.drop(columns=['issue_d', 'earliest_cr_line'], inplace=True)
```

### ***5. Converting to Numbers***

Few Columns in our dataset were string type having % value in them, but we require them to be in numerical format i.e. float64. So here we convert the columns `'int_rate', 'revol_util', 'bc_util', 'all_util', 'percent_bc_gt_75'` to float. We are not just removing the % sign and converting, we are making sure the impact remains the same so we divide the number obtained after removing the % sign by 100 to get a decimal value which is equivalent.

```python
# Removing % if it is present, converting value to number, and ensuring dti is numeric

pct_cols = ['int_rate', 'revol_util', 'bc_util', 'all_util', 'percent_bc_gt_75'] # Columns with %

for col in pct_cols:
    # Removing '%' if the column is stored as strings like "13.56%"
    neny_loan_clean[col] = (
        neny_loan_clean[col]
        .astype(str)
        .str.replace('%', '', regex=False)
    )

    # Converting to numeric
    neny_loan_clean[col] = pd.to_numeric(neny_loan_clean[col], errors='coerce')

    # Converting from 0–100 to 0–1 scale
    neny_loan_clean[col] = neny_loan_clean[col] / 100.0

neny_loan_clean['dti'] = pd.to_numeric(neny_loan_clean['dti'], errors='coerce')
```

### ***7. Manipulating the Raw Data***

#### **7.1. Numerical Variables**

Here we perform intelligent, rule-based imputation of missing values for our dataset by reconstructing variables using related fields, applying business logic, and then doing fallback median imputations.

```python
# Impute missing values Variable by Variable
df = neny_loan_clean.copy()

# 1. Reconstruct il_util where missing and inputs are available
if {'total_bal_il', 'total_il_high_credit_limit', 'il_util'}.issubset(df.columns):
    mask = df['il_util'].isna() & df['total_il_high_credit_limit'].gt(0)
    df.loc[mask, 'il_util'] = (
        df.loc[mask, 'total_bal_il'] / df.loc[mask, 'total_il_high_credit_limit']
    )

# 2. Reconstruct all_util from tot_cur_bal / tot_hi_cred_lim
if {'tot_cur_bal', 'tot_hi_cred_lim', 'all_util'}.issubset(df.columns):
    mask = df['all_util'].isna() & df['tot_hi_cred_lim'].gt(0)
    df.loc[mask, 'all_util'] = (
        df.loc[mask, 'tot_cur_bal'] / df.loc[mask, 'tot_hi_cred_lim']
    )

# 3. Reconstruct avg_cur_bal from tot_cur_bal / total_acc
if {'avg_cur_bal', 'tot_cur_bal', 'total_acc'}.issubset(df.columns):
    mask = df['avg_cur_bal'].isna() & df['total_acc'].gt(0)
    df.loc[mask, 'avg_cur_bal'] = (
        df.loc[mask, 'tot_cur_bal'] / df.loc[mask, 'total_acc']
    )

# 4. Reconstruct revol_util from revol_bal / total_rev_hi_lim
if {'revol_util', 'revol_bal', 'total_rev_hi_lim'}.issubset(df.columns):
    mask = df['revol_util'].isna() & df['total_rev_hi_lim'].gt(0)
    df.loc[mask, 'revol_util'] = (
        df.loc[mask, 'revol_bal'] / df.loc[mask, 'total_rev_hi_lim']
    )

# 5. Reconstruct bc_open_to_buy from total_bc_limit & bc_util
if {'bc_open_to_buy', 'bc_util', 'total_bc_limit'}.issubset(df.columns):
    mask = df['bc_open_to_buy'].isna() & df['total_bc_limit'].gt(0) & df['bc_util'].notna()
    # bc_util assumed in 0–1 units
    df.loc[mask, 'bc_open_to_buy'] = (
        df.loc[mask, 'total_bc_limit'] * (1 - df.loc[mask, 'bc_util'])
    )

# 6. Reconstruct bc_util from bc_open_to_buy & total_bc_limit
if {'bc_open_to_buy', 'bc_util', 'total_bc_limit'}.issubset(df.columns):
    mask = df['bc_util'].isna() & df['total_bc_limit'].gt(0) & df['bc_open_to_buy'].notna()
    df.loc[mask, 'bc_util'] = (
        1 - df.loc[mask, 'bc_open_to_buy'] / df.loc[mask, 'total_bc_limit']
    )

# 7. Handle structural zeros where there is no such product type

# If no bankcards, bankcard metrics equals 0
if {'num_bc_tl', 'bc_util', 'percent_bc_gt_75', 'bc_open_to_buy'}.issubset(df.columns):
    no_bc_mask = df['num_bc_tl'].eq(0) | df['num_bc_sats'].eq(0)
    df.loc[no_bc_mask, ['bc_util', 'percent_bc_gt_75', 'bc_open_to_buy']] = \
        df.loc[no_bc_mask, ['bc_util', 'percent_bc_gt_75', 'bc_open_to_buy']].fillna(0)

# If no installment accounts, il_util and mo_sin_old_il_acct equals 0
if {'num_il_tl', 'open_act_il', 'il_util', 'mo_sin_old_il_acct'}.issubset(df.columns):
    no_il_mask = df['num_il_tl'].eq(0) | df['open_act_il'].eq(0)
    df.loc[no_il_mask, ['il_util', 'mo_sin_old_il_acct']] = \
        df.loc[no_il_mask, ['il_util', 'mo_sin_old_il_acct']].fillna(0)

# num_tl_120dpd_2m: treat NaN as 0 (if there were severe delinquencies, they'd be reported)
if 'num_tl_120dpd_2m' in df.columns:
    df['num_tl_120dpd_2m'] = df['num_tl_120dpd_2m'].fillna(0)

# 8. Remaining missing values on variables: fallback imputation (median or conservative)

# revol_util & dti: fallback to median if still missing
for col in ['revol_util', 'dti', 'all_util', 'avg_cur_bal']:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)

# emp_title: treat missing as category "Unknown"
if 'emp_title' in df.columns:
    df['emp_title'] = df['emp_title'].fillna('Unknown')

neny_loan_clean = df
```

### ***Step A6 – Shared evaluation helper for all models***

`evaluate_model` is a generic helper that fits a model, retrieves `predict_proba` scores for the positive class, computes ROC AUC, applies a configurable decision threshold, and prints a confusion matrix plus classification report. Centralizing this logic keeps each model block small, readable, and consistent.


```python
def evaluate_model(name, model, X_train, y_train, X_test, y_test, threshold=0.5):
    """
    Generic evaluation helper for classifiers with predict_proba.
    Fits the model, computes AUC on probabilities, then generates a confusion
    matrix and classification report using a given threshold on predicted probs.
    """
    # Fit the model
    model.fit(X_train, y_train)

    # Predicted probabilities for positive class
    y_test_proba = model.predict_proba(X_test)[:, 1]

    # ROC AUC using probabilities
    auc = roc_auc_score(y_test, y_test_proba)

    # Convert probabilities to class predictions via threshold
    y_test_pred = (y_test_proba >= threshold).astype(int)

    print(f"\n=== {name} ===")
    print(f"ROC AUC (probabilities): {auc:.4f}")
    print(f"Threshold used: {threshold:.2f}")
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_test_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_test_pred, digits=4))

    return y_test_proba, y_test_pred, auc
```

#### ***Step A6.1. - ROC & AUC graph plotting helper***

We define `plot_roc_curve`, which takes true labels and predicted probabilities and plots the ROC curve along with its AUC. Each model reuses this helper so ROC plots share the same style and are easy to compare visually.


```python
def plot_roc_curve(y_true, y_proba, label, title_suffix=""):
    """
    Plot ROC curve for a given set of predicted probabilities.
    """
    fpr, tpr, _ = roc_curve(y_true, y_proba)
    auc_val = roc_auc_score(y_true, y_proba)

    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, label=f"{label} (AUC = {auc_val:.3f})")
    plt.plot([0, 1], [0, 1], linestyle="--", label="Random guess")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve – {label}{title_suffix}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

### ***Step L1 – Re‑check missing values in `X_train`***

### ***Step L2 – Define Logistic Regression pipeline***

We wrap preprocessing and the classifier into a single `Pipeline`:
- `SimpleImputer(strategy="median")` fills any remaining numeric gaps.
- `StandardScaler()` standardizes features to zero mean and unit variance.
- `LogisticRegression(max_iter=1000, class_weight="balanced")` provides a regularized linear model that can handle class imbalance.

Using a pipeline ensures identical preprocessing during training and evaluation, avoiding data‑leakage bugs.


```python
log_reg_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(
        max_iter=1000,
        n_jobs=-1,
        class_weight="balanced",   # handle imbalance
        penalty="l2",
        solver="lbfgs"
    ))
])

set_config(display='diagram')
log_reg_pipeline
```

```python
y_lr_proba, y_lr_pred, auc_lr = evaluate_model(
    name="Logistic Regression (baseline, threshold = 0.50)",
    model=log_reg_pipeline,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    threshold=0.5
)
```

### ***Step L4 – Hyperparameter tuning for Logistic Regression***

We use `RandomSearchCV` over different values of `C` (inverse regularization strength) while keeping the same preprocessing pipeline. The search is scored with ROC AUC using 3‑fold cross‑validation, giving a slightly better‑calibrated linear model without overfitting.


```python
# We refine the model a bit by tuning C (inverse regularization strength).
# Lower C = stronger regularization, higher C = more flexible model.

log_reg_base = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(
        max_iter=1000,
        n_jobs=-1,
        class_weight="balanced",
        penalty="l2",
        solver="lbfgs"
    ))
])

# Use a random search over C instead of full grid search
param_distributions_lr = {
    "clf__C": uniform(loc=0.01, scale=10.0)
}

rand_lr = RandomizedSearchCV(
    estimator=log_reg_base,
    param_distributions=param_distributions_lr,
    n_iter=20,
    scoring="roc_auc",
    n_jobs=-1,
    cv=3,
    verbose=1,
    random_state=42
)

rand_lr.fit(X_train, y_train)

print("Best LR AUC (CV):", rand_lr.best_score_)
print("Best params:", rand_lr.best_params_)

log_reg_tuned = rand_lr.best_estimator_
```

## **Phase X – XGBoost**

XGBoost is a powerful gradient‑boosted tree ensemble that often performs very well on structured tabular data. We first build a reasonably configured baseline model using a `scale_pos_weight` value derived from the observed class imbalance.


### ***Step X1 – Compute `scale_pos_weight` from imbalance***

For XGBoost we compute `scale_pos_weight = #neg / #pos` from the training data. This tells the booster to pay more attention to the minority (default) class without needing to resample the data.


```python
neg_count = (y_train == 0).sum()
pos_count = (y_train == 1).sum()
scale_pos_weight = neg_count / pos_count

print("neg_count:", neg_count)
print("pos_count:", pos_count)
print("scale_pos_weight:", scale_pos_weight)
```

### ***Step X2 – Define baseline XGBoost model***

We instantiate `XGBClassifier` with a binary logistic objective, `eval_metric="logloss"`, histogram‑based tree method, and reasonable defaults for depth, learning rate, subsampling, and `scale_pos_weight`. This forms our first XGBoost benchmark.


```python
xgb_baseline = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    tree_method="hist",
    random_state=42,
    n_estimators=400,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight
)
```

### ***Step X3 – Evaluate baseline XGBoost (threshold = 0.5)***

Using the shared `evaluate_model` helper, we fit the baseline XGBoost model and inspect its AUC, confusion matrix, and classification report at a 0.5 threshold.


```python
y_xgb_base_proba, y_xgb_base_pred, auc_xgb_base = evaluate_model(
    name="XGBoost (baseline, threshold = 0.50)",
    model=xgb_baseline,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    threshold=0.5
)
```

### ***Phase X4 – XGBoost hyperparameter tuning***

We now tune XGBoost using a randomized search over tree depth, number of estimators, learning rate, subsampling, and column‑sampling parameters. The goal is to push performance a bit higher while avoiding overfitting.


#### ***Step X4.1. – Define base model and search space***

We set up a base `XGBClassifier` and specify reasonable ranges for key hyperparameters (for example `n_estimators`, `max_depth`, `subsample`, `colsample_bytree`). These ranges define the space that `RandomizedSearchCV` will explore.


```python
xgb_base_tune = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    tree_method="hist",
    random_state=42,
    n_estimators=400,
    learning_rate=0.05,
    scale_pos_weight=scale_pos_weight
)

param_distributions = {
    "max_depth": randint(3, 8),
    "min_child_weight": randint(1, 6),
    "subsample": uniform(0.6, 0.4),        # 0.6–1.0
    "colsample_bytree": uniform(0.6, 0.4), # 0.6–1.0
    "gamma": uniform(0, 5),
    "reg_lambda": uniform(0.5, 3.0)
}
```

#### ***Step X4.2. – Run `RandomizedSearchCV` on the training set***

We run `RandomizedSearchCV` for a limited number of iterations, optimizing ROC AUC with 3‑fold cross‑validation. This is more computationally efficient than a full grid search while still exploring many hyperparameter combinations.


```python
xgb_search = RandomizedSearchCV(
    estimator=xgb_base_tune,
    param_distributions=param_distributions,
    n_iter=30,
    scoring="roc_auc",
    n_jobs=-1,
    cv=3,
    verbose=1,
    random_state=42
)

xgb_search.fit(X_train, y_train)

print("Best XGB CV AUC:", xgb_search.best_score_)
print("Best params:", xgb_search.best_params_)

best_xgb_params = xgb_search.best_params_
```

#### ***Step X4.3. – Build tuned XGBoost with best parameters***

Using the best hyperparameters found by the randomized search, we construct `xgb_tuned`. This is our final tuned gradient‑boosted tree model.


```python
xgb_tuned = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    tree_method="hist",
    random_state=42,
    n_estimators=400,
    learning_rate=0.05,
    scale_pos_weight=scale_pos_weight,
    **best_xgb_params
)
```

##### ***Step X4.4.1. – ROC curve for tuned XGBoost***

We re‑run `evaluate_model` (for logging) and then call `plot_roc_curve` to visualize the ROC curve of the tuned XGBoost model. We will reuse these probabilities in the final combined ROC comparison.


```python
y_xgb_tuned_proba, y_xgb_tuned_pred, auc_xgb_tuned = evaluate_model(
    name="XGBoost (tuned, threshold = 0.50)",
    model=xgb_tuned,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    threshold=0.5
)

plot_roc_curve(y_test, y_xgb_tuned_proba, label="XGBoost (tuned)")
```

### ***Step X5.1. – Rebuild the threshold sweep*** (XGB New Threshold Test)

Instead of fixing the threshold at 0.5, we sweep thresholds from 0.05 to 0.95. For each threshold we recompute precision, recall, F1‑score, and confusion‑matrix components for the tuned XGBoost model and store the results in `thr_df`.


```python
#Define thresholds to test
thresholds = np.linspace(0.05, 0.95, 19)  # 0.05, 0.10, ..., 0.95

rows = []

for thr in thresholds:
    # Convert probabilities to class predictions at this threshold
    y_pred_thr = (y_xgb_tuned_proba >= thr).astype(int)

    # Metrics for the default class (1)
    precision = precision_score(y_test, y_pred_thr, pos_label=1, zero_division=0)
    recall = recall_score(y_test, y_pred_thr, pos_label=1, zero_division=0)
    f1 = f1_score(y_test, y_pred_thr, pos_label=1, zero_division=0)

    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_thr).ravel()

    rows.append({
        "threshold": thr,
        "precision_1": precision,
        "recall_1": recall,
        "f1_1": f1,
        "tn": tn,
        "fp": fp,
        "fn": fn,
        "tp": tp
    })

thr_df = pd.DataFrame(rows)

# Quick sanity check
print(thr_df.head())
print(thr_df.columns)
```

### ***Step X5.2. – Choose best threshold (max F1 for default class)***

We select the threshold that maximizes F1‑score for the positive (default) class. This gives a decision rule that balances precision and recall according to F1 instead of using the arbitrary 0.5 cutoff.


```python
best_row = thr_df.loc[thr_df["f1_1"].idxmax()]
best_threshold = best_row["threshold"]
best_row
```

### ***Step X5.3. - Final evaluation at chosen threshold***

Finally, we apply the best threshold to the tuned XGBoost probabilities and recompute AUC, confusion matrix, and classification report. This is effectively our “production‑style” evaluation of the chosen model and threshold.


```python
# Reproduces: AUC: 0.7345, Confusion matrix at threshold 0.70, Precision/recall/F1 you already observed.
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

y_xgb_final_pred = (y_xgb_tuned_proba >= best_threshold).astype(int)
auc_final = roc_auc_score(y_test, y_xgb_tuned_proba)

print(f"Final XGBoost ROC AUC (probabilities): {auc_final:.4f}")
print(f"Chosen threshold: {best_threshold:.2f}")

print("\nConfusion Matrix (final threshold):")
print(confusion_matrix(y_test, y_xgb_final_pred))

print("\nClassification Report (final threshold):")
print(classification_report(y_test, y_xgb_final_pred, digits=4))

# Produces: Confusion matrix at threshold 0.70 and Precision/recall/F1 you already observed.
```

## **Comparison of tuned models**

To summarize the performance of all tuned models, we plot their ROC curves on a single figure using their predicted probabilities on the test set (`y_test`). This highlights which model achieves the best trade-off between true-positive and false-positive rates and provides a clear visual justification for the final model choice.


### ***Combined RoC AUC Graph***

```python
models_for_roc = [
    ("Logistic Regression (tuned)", y_lr_tuned_proba),
    ("Decision Tree (tuned)",      y_dt_tuned_proba),
    ("Random Forest (tuned)",      y_rf_tuned_proba),
    ("XGBoost (tuned)",            y_xgb_tuned_proba),
]

plt.figure(figsize=(7, 7))

for label, y_proba in models_for_roc:
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc_val = roc_auc_score(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{label} (AUC = {auc_val:.3f})")

# Reference line for random guessing
plt.plot([0, 1], [0, 1], linestyle="--", label="Random guess")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison – Tuned Models")
plt.legend()
plt.grid(True)
plt.show()
```

### ***Comparison of All the Trained Modles***

```python
# Dictionary of tuned models' predictions and probabilities
tuned_results = {
    "Logistic Regression (tuned)": {
        "y_pred":  y_lr_tuned_pred,
        "y_proba": y_lr_tuned_proba,
    },
    "Decision Tree (tuned)": {
        "y_pred":  y_dt_tuned_pred,
        "y_proba": y_dt_tuned_proba,
    },
    "Random Forest (tuned)": {
        "y_pred":  y_rf_tuned_pred,
        "y_proba": y_rf_tuned_proba,
    },
    "XGBoost (tuned)": {
        "y_pred":  y_xgb_tuned_pred,
        "y_proba": y_xgb_tuned_proba,
    },
}

rows = []

for model_name, res in tuned_results.items():
    y_pred = res["y_pred"]
    y_proba = res["y_proba"]

    # Full classification report as a dict
    report = classification_report(y_test, y_pred, output_dict=True)

    # Use the positive/default class ("1"); fall back safely if needed
    pos_label = "1"
    if pos_label not in report:
        class_keys = [k for k in report.keys() if k.isdigit()]
        pos_label = sorted(class_keys)[-1]  # last numeric label

    pos_metrics    = report[pos_label]
    macro_avg      = report["macro avg"]
    weighted_avg   = report["weighted avg"]
    auc            = roc_auc_score(y_test, y_proba)

    rows.append({
        "Model": model_name,

        # Positive class metrics (Precision, Recall, F1, Support)
        f"Precision (class {pos_label})": pos_metrics["precision"],
        f"Recall (class {pos_label})":    pos_metrics["recall"],
        f"F1-score (class {pos_label})":  pos_metrics["f1-score"],
        f"Support (class {pos_label})":   pos_metrics["support"],

        # ROC AUC based on probabilities
        "ROC AUC": auc,

        # Macro Average (unweighted)
        "Precision (macro avg)": macro_avg["precision"],
        "Recall (macro avg)":    macro_avg["recall"],
        "F1-score (macro avg)":  macro_avg["f1-score"],

        # Weighted Average (by support)
        "Precision (weighted avg)": weighted_avg["precision"],
        "Recall (weighted avg)":    weighted_avg["recall"],
        "F1-score (weighted avg)":  weighted_avg["f1-score"],
    })

comparison_df = pd.DataFrame(rows)
comparison_df.set_index("Model").round(4)
```

##**Conclusion**

Across the four tuned models evaluated on the held-out test set—Logistic Regression, Decision Tree, Random Forest, and XGBoost—the **XGBoost model emerged as the best overall performer**. It achieved the **highest ROC AUC of 0.7345** (vs. 0.7305 for Logistic Regression, 0.7233 for Random Forest, and 0.7015 for the Decision Tree) and the **highest weighted F1-score of 0.7944** (compared with 0.7631, 0.7433, and 0.6050, respectively). At the aggregate level, all models show strong weighted precision around **0.95**, but XGBoost offers the best balance of weighted precision and recall (0.9508 / 0.6986), indicating fewer overall misclassifications once class imbalance is taken into account.

However, the class-1 (default) metrics highlight the intrinsic difficulty of predicting rare defaults. Even for XGBoost, **precision for defaulters is only 0.0693 while recall is 0.6332 (F1 = 0.1249)**, meaning the model successfully captures a majority of true defaulters but also flags many borrowers who ultimately repay. This pattern is consistent across models and reflects the very low base default rate in the dataset. The practical interpretation is that the final XGBoost model is most useful as a **risk-ranking and early-warning tool**—to prioritize manual review, adjust pricing, or trigger proactive monitoring—rather than as a fully automated approve/decline engine. Used in this way, the model can meaningfully support stakeholders’ credit risk decisions while recognizing data limitations (regional scope, 2019–2020 window, class imbalance) and the need for ongoing monitoring and threshold tuning.
